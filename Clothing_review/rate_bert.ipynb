```python
from utils import clean_cr_dataset_on_Rating, clean_cr_dataset_on_Recommended_IND, pre_proccess_data
from BERT import BERT
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification


tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
data = pd.read_csv('./w_reviews.csv', index_col=0)
data = pre_proccess_data(data)

data_ra = clean_cr_dataset_on_Recommended_IND(data)
data_ra = data_ra.sample(n=100)
```

```python
bert_ra = BERT(tokenizer = tokenizer, model = model, data = data_ra, target="Text", evaluation_target='Recommended IND', target_type="sentiment")
bert_ra.simple_prediction()
bert_ra.data.to_csv("data_set_con_prediccion.csv")
```

```python
data_s = data.sample(1000)
dta_s_bert = BERT(tokenizer = tokenizer, model = model, data = data_s, target="Text", evaluation_target='Recommended IND', target_type="sentiment")
dta_s_bert.simple_prediction()
dta_s_bert.data.to_csv("data_set_con_prediccion1000.csv")
```
